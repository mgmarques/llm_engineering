## Types of prompts

You may know this already - but if not, you will get very familiar with it!

Models like GPT have been trained to receive instructions in a particular way.

They expect to receive:

**A system prompt** that tells them what task they are performing and what tone they should use

**A user prompt** -- the conversation starter that they should reply to

````python
# Define our system prompt - you can experiment with this later, changing the last sentence to 'Respond in markdown in Spanish."

system_prompt = """
You are a snarky assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
````

````python
# Define our user prompt

user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
````

## Messages

The API from OpenAI expects to receive messages in a particular structure.
Many of the other APIs share this structure:

```python
[
    {"role": "system", "content": "system message goes here"},
    {"role": "user", "content": "user message goes here"}
]
```
To give you a preview, the next 2 cells make a rather simple call - we won't stretch the mighty GPT (yet!)

````python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "What is 2 + 2?"}
]

response = ollama.chat.completions.create(model="llama3.2", messages=messages)
response.choices[0].message.content
````

## And now let's build useful messages for GPT-4.1-mini, using a function

````python
# See how this function creates exactly the format above

def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]
````

````python
# Try this out, and then try for a few more websites

messages_for(ed)
````

## Time to bring it together - the API for OpenAI is very simple!

````python
# And now: call the OpenAI API. You will get very familiar with this!

def summarize(url):
    website = fetch_website_contents(url)
    response = ollama.chat.completions.create(
        model = "llama3.2",
        messages = messages_for(website)
    )
    return response.choices[0].message.content
````

````python
summarize("https://edwarddonner.com")
````

````python
# A function to display this nicely in the output, using markdown

def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
````

````python
# Step 1: Create your prompts

system_prompt = """
You are a snarky assistant that analyzes the contents of a email, 
and will suggest an appropriate short subject line for the email.
If it includes news or announcements, then do not consider those.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
user_prompt = """
Here are the contents of a email.
Provide a short subject line for the email.
If it includes news or announcements, then do not consider those.
"""

email_contents = """
Estamos entrando em contato para informar que atualizamos os Termos de Uso da HBO Max (“Termos”). Essas atualizações incluem alterações relacionadas, por exemplo, às informações sobre nossas ofertas de assinatura, áreas de disponibilidade internacional do serviço e informações sobre sua conta.

Você pode consultar os Termos atualizados aqui. Esses Termos atualizados entrarão em vigor imediatamente para novos assinantes. Para assinantes existentes como você, estes Termos atualizados entrarão em vigor dia 16 de abril de 2026.

Se você tem uma assinatura ativa, a manutenção de sua assinatura e/ou uso da HBO Max a partir de 16 de abril de 2026, confirma que você revisou e concordou com os Termos atualizados.

Agradecemos por fazer parte da comunidade HBO Max!

Precisa de ajuda? Visite nosso Centro de Ajuda.
"""

# Step 2: Make the messages list
def messages_for(email_content):
    return [
        {"role": "system", "content": system_prompt + email_content},
        {"role": "user", "content": user_prompt + email_content}
    ]

# Step 3: Call OpenAI
# response =
def titlize(email_content):
    response = ollama.chat.completions.create(
        model = "llama3.2",
        messages = messages_for(email_content)
    )
    return response.choices[0].message.content

# Step 4: print the result
def display_title(email_contents):
    display(Markdown(titlize(email_contents)))

display_title(email_contents)
````

# Radio Calico - Song Details Summarizer Prototype

## Overview

This prototype is a **music information assistant** that takes song details from Radio Calico (a music streaming/radio service) and generates rich, formatted summaries using a local LLM (Llama 3.2) via Ollama.

## How It Works

### Architecture

```
Radio Calico song metadata → Prompt construction → Ollama (Llama 3.2) → Markdown response → Jupyter display
```

1. **Input**: Song details string containing track name, artist, album, and album cover URL (as broadcast by Radio Calico)
2. **LLM Backend**: Uses Ollama running locally (`localhost:11434`) with OpenAI-compatible API via the `openai` Python SDK
3. **Output**: Rendered Markdown in a Jupyter notebook

### Components

| Component | Purpose |
|---|---|
| `system_prompt` | Instructs the LLM to act as a genre-aware music enthusiast. Directs it to extract metadata, look up facts, provide lyrics, album tracklists, and related products. Strips the "Radio Calico" source attribution. |
| `user_prompt` | The query template — asks for song meaning, lyrics, facts, album tracklist table, related products, and a funny multilingual closing note. |
| `messages_for(song_details, language)` | Builds the chat message array, appending the target language to the system prompt and song details to the user prompt. |
| `summarize(song_details, language)` | Sends the messages to Ollama's `llama3.2` model and returns the raw text response. |
| `display_summary(song_details, language)` | Orchestrates the call and renders the result as Markdown in Jupyter. |

### Configuration

- **LLM**: Llama 3.2 via Ollama (local, no API key needed — `api_key="anything"` is a placeholder)
- **Language**: Configurable per call — supports `"English"`, `"Portuguese"`, `"Spanish"`, etc.
- **Model swap**: Change the `model` parameter in `summarize()` to use any Ollama-hosted model

### Example Input

```
Listening to "Under the Milky Way" by The Church (Starfish (Deluxe Edition)) on Radio Calico!
Album cover: https://is1-ssl.mzstatic.com/image/thumb/Music114/v4/.../300x300bb.jpg
```

### Expected Output

A Markdown document containing:
- Song description/meaning
- Lyrics
- Interesting facts about the song, album, and artists
- Album tracklist table with durations
- Related products summary (if applicable)
- A humorous closing quote in a rotating language

## Limitations & Notes

- **Lyrics accuracy**: LLMs hallucinate lyrics — the model generates from training data, not a lyrics database. For production, integrate a lyrics API (e.g., Musixmatch, Genius).
- **Album cover analysis**: The system prompt mentions checking the album cover URL/image metadata, but the current implementation passes it as text only — the model cannot actually fetch or view the image. A multimodal model or a separate image-fetch step would be needed.
- **Local-only**: Requires Ollama running on the same machine. For production deployment, consider a hosted LLM or containerized Ollama.
- **No error handling**: Network failures, Ollama downtime, or malformed input will raise unhandled exceptions.

## Suggested Improvements for Production (Radio Calico)

1. **Add error handling** around the Ollama API call
2. **Integrate a real lyrics API** instead of relying on LLM-generated lyrics
3. **Use a multimodal model** (or fetch the album cover separately) if image analysis is needed
4. **Cache responses** — the same song doesn't need to be summarized twice
5. **Parameterize the model name** instead of hardcoding `"llama3.2"`
6. **Structured output** — consider returning JSON for programmatic consumption alongside Markdown for display

````python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display


system_prompt = """
You are a music enthusiast who uses language appropriate to the genre kind of music being evaluated.
Ignore that is from Radio Calico
Check any metaddata you can find in the song details, such as the album name, artists, and song name, and use this to inform your response.
Check the album cover URL, the metadata of its image and use this to inform your response.
If asked, give a brief summary of interesting facts about:
 * The song
 * The album
 * The artists
If asked, give a table containing the informations about its ablbum, production year, record label, and if the list of songs with its genere and durations.
If asked, give a summary of the products related to the song, album or artists.
If asked, give the song lirycs
If asked anything else, ignore it.
Make a Radio Calico Merchandise as a Nerd Shirt with the album cover and the name of the song.
Finish with a "last one phrase" that seems to have been said within the song itself

Respond just with the markdown and images
Respond on 
"""

user_prompt = """
Here are the song details.
Tell what is about?
Give me the song lirycs
Give me interesting facts about the song, album and artists?
Give a table containing the informations about its ablbum if the list of songs and its durations.
If it includes products, then summarize these too.
Finally, just at the end, say something funny as a closing note in other languages at each time
"""


OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="anything")

def messages_for(song_details, language):
    return [
        {"role": "system", "content": system_prompt + language},
        {"role": "user", "content": user_prompt + song_details}
    ]

def summarize(song_details, language):
    # Create a new set of messages for the OpenAI chat
    messages = messages_for(song_details, language=language)

    # Create a new message and ask OpenAI to generate a summary
    response = ollama.chat.completions.create(model="llama3.2", messages=messages)

    return response.choices[0].message.content

def display_summary(song_details, language):
    summary = summarize(song_details, language)
    display(Markdown(summary))

song_details = """
Listening to "Under the Milky Way" by The Church (Starfish (Deluxe Edition)) on Radio Calico!

Album cover: https://is1-ssl.mzstatic.com/image/thumb/Music114/v4/a4/87/86/a48786f3-5c92-30d6-be13-de7b030d6d7e/dj.mhnucynb.jpg/300x300bb.jpg
"""
language =  "English" #"Portuguese" "Spanish"
display_summary(song_details=song_details, language=language)
````

````python
"""
When I try https://openai.com it doesn't work! That's because OpenAI has a fancy website that uses Javascript. 
So, we use Selenium to runs a browser behind the scenes, renders the page, and allows me to query it. 
"""
#!pip install selenium beautifulsoup4 

import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
from selenium import webdriver
from bs4 import BeautifulSoup

system_prompt = """
You are an ai assistant that analyzes the contents of a website,
and provides a short, technical, and good humored summary with a spech as a smart guy a little excited.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""

user_prompt = """
Here are the contents of an AI website.
Tell what is about this website!
What is the main topic of this website?
Provide a short summary of what this website is about.
If it includes products, then summarize these too.
Give me the hierarchy navigation of the website, with all sections and subsections, and a short summary of each of them.
Finally, just at the end, say something funny as a closing note in other languages at each time
"""


OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="anything")

def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]

def summarize(url):
    # Create a new instance of the Chrome driver
    options = webdriver.ChromeOptions()
    options.add_argument('headless')  # Run in headless mode
    driver = webdriver.Chrome(options=options)

    # Navigate to the URL
    driver.get(url)

    # Parse HTML and extract text content
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    text_content = ''
    for element in soup.find_all(TextDisplayObject=True):
        text_content += element.strip()

    driver.quit()  # Close the browser

    # Create a new set of messages for the OpenAI chat
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + text_content}
    ]

    # Create a new message and ask OpenAI to generate a summary
    response = ollama.chat.completions.create(model="llama3.2", messages=messages)

    return response.choices[0].message.content

def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))

display_summary("https://openai.com/")
````
