## Support Methods

`````python
# Support Methods

def set_debate() -> None:
    """
    Sets up a debate scenario with three participants, a topic, and a conversational flow.

    This function initializes the core elements of a debate setup, including the participants,
    their speech personalities, the debate topic, and an initial conversational prompt.
    It prepares the initial conversation string for display.
    
    Side Effects:
        - Prints progress messages to stdout.
        - Sets global variables to configure the debate.

    Dependencies:
        - random
        - openai (assumed to be configured with API key)
    """
    global participants, one, two, three, one_speech_personality, two_speech_personality, three_speech_personality, \
           topic, emotional, outside, atmosphere, conversation, focus
    participants = random.sample(Characters, k=3)
    one = list(participants[0].keys())[0]
    two = list(participants[1].keys())[0]
    three = list(participants[2].keys())[0]

    one_speech_personality = list(participants[0].values())[0]
    two_speech_personality = list(participants[1].values())[0]
    three_speech_personality = list(participants[2].values())[0]

    # Set the main topic
    topic = random.choices(topics, k=1)[0]

    # Define the Emotional/Social tone
    emotional = random.choices(emotions, k=1)[0]

    # choice the main Narrative/Storytelling
    narrative = random.choices(narratives, k=1)[0]

    # Establish the Environmental/Physical or Cultural/Architectural 
    outside = random.choices(outsides, k=1)[0]

    atmosphere = f"{emotional} dominate by the narrative {narrative}, impacted by the {outside} environmental"

    focus = f"{topic} where {atmosphere}"

    begining =f"""
    Let's watch a debate about the {topic} the main atmosphere is {atmosphere}.\n 
    The participants in today's debate are:
    * {one}: {one_speech_personality}
    * {two}: {two_speech_personality} 
    * {three}: {three_speech_personality}
    
    So {one} how you wat to start with {focus}
    """
    conversation =[]
    conversation.append(begining)

    display(Markdown(f"# Three are debating on the dome, one man leaves?\n{begining.format(Markdown)}\n"))

def set_jury() -> None:
    """
    Sets up a jury for a trial scenario, assigning jury members and their speech personalities.

    This function creates the jury panel with three members, each with a distinct
    speech personality, and prepares a statement to be displayed, outlining their roles.

    Side Effects:
        - Prints progress messages to stdout.
        - Sets global variables to configure the trial.

    Dependencies:
        - random
        - openai (assumed to be configured with API key)
        - markdown (assumed to be installed)
    """
    global trial, jury, jury_one, jury_two, jury_three, jury_one_speech_personality, jury_two_speech_personality, jury_three_speech_personality, votes

    rest = [character for character in Characters if character not in participants]
    jury_members = random.sample(rest, k=3)

    jury_one = list(jury_members[0].keys())[0]
    jury_two = list(jury_members[1].keys())[0]
    jury_three = list(jury_members[2].keys())[0]

    jury_one_speech_personality = list(jury_members[0].values())[0]
    jury_two_speech_personality = list(jury_members[1].values())[0]
    jury_three_speech_personality = list(jury_members[2].values())[0]

    jury = f"""
    Here the jury members:
    * {jury_one}: {jury_one_speech_personality}
    * {jury_two}: {jury_two_speech_personality} 
    * {jury_three}: {jury_three_speech_personality}"

    From the deabte above you must choose only one who wins from:\n
    * {one}
    * {two}
    * {three}
    Take into account the ability to convince, entertain, persuade, and engage the public! 
    """
    trial = [f"{jury}\n{conversation}"]
    votes = []

    display(Markdown(f"## The trial!\n{jury.format(Markdown)}\n"))    

def llm_debate_configuration(the_floor: str, the_floor_personality: str, opponent_1: str, opponent_2: str)-> Tuple[str, str]:
    """
    Configures the system and user prompts for an LLM to simulate a debate.

    This function crafts the prompts that will be sent to a Large Language Model (LLM)
    to generate the dialogue for the debate. It specifies the roles of the participants
    and the overall context of the conversation.

    Args:
        the_floor (str): The name and personality of the speaker in the debate.
        the_floor_personality (str): The personality of the speaker.
        opponent_1 (str): The name of the first opponent.
        opponent_2 (str): The name of the second opponent.

    Returns:
        Tuple[str, str]: A tuple containing the system prompt and the user prompt for the LLM.
    """
    system_prompt = f"""
    You are {the_floor} with a personality: {the_floor_personality}. 
    You are in a conversation with {opponent_1} and {opponent_2} about {topic}."""

    user_prompt = f"""
    You are {the_floor}, in conversation with with {opponent_1} and {opponent_2} about {topic} the main atmosphere is {atmosphere}.
    The conversation so far is as follows:
    {conversation}
    Now it's your turn, answer with what you would like to say next as {the_floor}, limited to two paragraphs if you reply to someone, 
    otherwise limited to only one paragraph 
    Do not add a title with {the_floor} or mention it.
    """
    
    return system_prompt, user_prompt

def llm_trial_configuration(main_jury: str, main_jury_personality: str) -> Tuple[str, str]:
    """
    Configures the system and user prompts for an LLM to simulate a trial.

    This function crafts the prompts that will be sent to a Large Language Model (LLM)
    to generate the dialogue for the trial scenario. It specifies the role of the
    main jury member and the overall context of the conversation.

    Args:
        main_jury (str): The name and personality of the main jury member.
        main_jury_personality (str): The personality of the main jury member.

    Returns:
        Tuple[str, str]: A tuple containing the system prompt and the user prompt for the LLM.
    """
    system_prompt = f""""
    You are {main_jury} with a personality: {main_jury_personality}. 
    You are in a trial and need vote on only one winer from {one}, or {two} or {three}."""

    user_prompt = f"""
    You are one of the members from a trial about {topic}
    {trial}
    You are {main_jury}, and now it's your turn, make your vote as {main_jury} on only one winer from {one}, or {two} or {three}.
    You are limited to only one paragraph. 
    Do not add a title with {main_jury} or mention it.
    Choose only one winner from {one}, or {two} or {three}!.
    You can not vote on any other character/person, do not vote for yourself, either.
    At the end, express clearly how is the winner from {one}, or {two} or {three}, said "My vote gos to: name of the winner.
    """

    return system_prompt, user_prompt

def turn_to_speak(speaker: str, client: OpenAI, llm_model: str) -> None:
    """
    Use LLM to present a speech in the debate.

    Args:
        speaker (str): The speaker.
        client (OpenAI): An OpenAI client object.
        llm_model (str): The name of the OpenAI model to use.

    Returns:
        None
    """
    global conversation
    stream = client.chat.completions.create(
        model=llm_model, 
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
            ],
        stream=True)

    response = f"---\n### {speaker}:\n"
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
    
    conversation.append(response)


def jury_trial(speaker: str, client: OpenAI, llm_model: str) -> str:
    """
    Uses the LLM to provide a final judgment based on the votes.

    Args:
        speaker (str): The speaker.
        client (OpenAI): An OpenAI client object.
        llm_model (str): The name of the OpenAI model to use.

    Returns:
        str: The response from the LLM.
    """
    global votes
    stream = client.chat.completions.create(
        model=llm_model, 
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
            ],
        stream=True)

    response = f"---\n### {speaker}:\n "
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)

    return response


def final_judgment(client: OpenAI, llm_model: str, votes: list[str]) -> None:
    """
    Uses the LLM to provide a final judgment based on the votes.

    Args:
        client (OpenAI): An OpenAI client object.
        llm_model (str): The name of the OpenAI model to use.
        votes (list[str]): List of votes for the debate.

    Returns:
        None
    """
    stream = client.chat.completions.create(
        model=llm_model, 
        messages=[
            {"role": "system", "content": "A maxter judge"},
            {"role": "user", "content": f"""
             According to that votes how is the winner from {one}, {two} or {three}?
              * The jury members only have one vote each, look for his last final vote expresed like "My vote goes to"!
              * The max of total count votes is 3.
              * The winner will be the one who receives the most votes.
             ## Votes:
             {votes}"""}
            ],
        stream=True)

    response = f"---\n## Final Judgemt:\n"
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
`````

## Main logic of the debate and voting

`````python

def have_fan() -> None:
    """
    This function simulates a debate between three entities (one, two, and three) with a jury.
    It orchestrates a series of turns, using LLMs with different personalities and models, 
    and then culminates in a final judgment by the jury.

    The function performs the following steps:
    1. Initializes the debate setup by calling `set_debate()`.
    2. Executes three rounds of debate, where each entity takes a turn arguing, 
       using different LLM configurations and speech personalities.  The turns are managed
       using `llm_debate_configuration` and `turn_to_speak`.
    3. Establishes a jury by calling `set_jury()`.
    4. Conducts a trial with the jury, again using different LLM configurations 
       and speech personalities for the jury members.  The jury's decision is obtained
       using `jury_trial`.
    5. Accumulates the jury's votes into the `votes` list.
    6.  Finally, it delivers a final judgment based on the collected votes, 
        using a designated model and router. 

    Raises:
        None explicitly. Any errors related to LLM calls, set-up or the trial process may occur. 

    Side Effects:
        - Prints progress messages to stdout during the debate rounds.
        - Appends votes to the `votes` global variable.
        - Performs a final judgment, potentially logging this information.
    """
    global system_prompt, user_prompt, votes
    set_debate()

    for i in range(3):
        system_prompt, user_prompt = llm_debate_configuration(one, one_speech_personality, two, three)
        turn_to_speak(one, ollama, LAMMA_1B_MODEL)
        system_prompt, user_prompt = llm_debate_configuration(two, two_speech_personality, one, three)
        turn_to_speak(two, ollama, LAMMA_MODEL)
        system_prompt, user_prompt = llm_debate_configuration(three, three_speech_personality, two, one)
        turn_to_speak(three, ollama, LAMMA_GEMMA)

    set_jury()
    system_prompt, user_prompt = llm_trial_configuration(jury_one, jury_one_speech_personality)
    vote = jury_trial(jury_one, ollama, LAMMA_1B_MODEL)
    votes.append(vote)
    system_prompt, user_prompt = llm_trial_configuration(jury_two, jury_two_speech_personality)
    vote = jury_trial(jury_two, openrouter, GEMINI_ROUTER)
    votes.append(vote)
    system_prompt, user_prompt = llm_trial_configuration(jury_three, jury_three_speech_personality)
    vote = jury_trial(jury_three, openrouter, GPT_ROUTER)
    votes.append(vote)

    # Final result:
    final_judgment(ollama, LAMMA_MODEL, votes)
`````
